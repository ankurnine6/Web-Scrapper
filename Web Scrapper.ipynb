```python
import requests
from bs4 import BeautifulSoup
import time
from tqdm import tqdm
import json
```

```python
hotel_name = []
link = []
num_review = []
price = []
rate = []
check_in = []
check_out = []
addr = []
room = []
hotel_info = []

url = 'https://me.cleartrip.com/hotels/united-states/miami/'
r = requests.get(url)
soup = BeautifulSoup(r.text, 'lxml')

for i in range (len(soup.findAll('h2',{"class":"hotels-name"}))) :

    hotel_name.append(soup.findAll('h2',{"class":"hotels-name"})[i].span.string)
    print("hotel name : "+ hotel_name[i])

    link.append(soup.findAll('h2',{"class":"hotels-name"})[i].findAll('a')[0]['href'])
    print('hotel_url : https://me.cleartrip.com/'+link[i])
    
    rate.append(soup.findAll("span",{'class':'hotelMeta'})[i].findAll("span")[0]["title"])
    print("hotel rating  : "+ rate[i])
    
    num_review.append(soup.findAll("span",{'class':'taReviews'})[i].text)
    print("hotel_review_num  : " + num_review[i])
    
    price.append(soup.findAll("div",{'class':'text-right book-now'})[i].text)
    print('hotel_price : ' +price[i])    
    
for i in range(2,39):
    url = 'https://me.cleartrip.com/hotels/united-states/miami?page='+str(i)
    print("page"+str(i))
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'lxml')
    for j in range(len(soup.findAll('h2',{'class':"hotels-name"}))):
        hotel_name.append(soup.findAll('h2',{"class":"hotels-name"})[j].span.string)
        print("hotel name : "+ soup.findAll('h2',{"class":"hotels-name"})[j].span.string)
        
        link.append(soup.findAll('h2',{'class':"hotels-name"})[j].findAll('a')[0]['href'])
        print('hotel_url : https://me.cleartrip.com/'+soup.findAll('h2',{'class':"hotels-name"})[j].findAll('a')[0]['href'])
        
        rate.append(soup.findAll("span",{'class':'hotelMeta'})[j].findAll("span")[0]["title"])
        print("hotel rating  : "+ soup.findAll("span",{'class':'hotelMeta'})[j].findAll("span")[0]["title"])
        
        try :
            num_review.append(soup.findAll("span",{'class':'taReviews'})[j].text)
            print("hotel_review_num  : " +soup.findAll("span",{'class':'taReviews'})[j].text)
            
        except :
            num_review.append("nan")
            print("no review")
            
            
    
        price.append(soup.findAll("div",{'class':'text-right book-now'})[j].text)
        print('hotel_price : ' +soup.findAll("div",{'class':'text-right book-now'})[j].text)
    
    
```

```python
hotel_info = []
for l in range(len(link)):
    url_new = 'https://me.cleartrip.com'+link[l]
    
    r = requests.get(url_new)
    soup_new = BeautifulSoup(r.text, 'lxml')
    

    try:
        addr.append(soup_new.findAll('small')[3].text.replace('\t','').replace('\n',''))
        print("hotel address  : "+ soup_new.findAll('small')[3].text.replace('\t','').replace('\n',''))

    except :
        addr.append("nan")
        print("hotel address  : nan")

    try:
        check_in.append(soup_new.findAll("div",{"class":"amenitiesCategory"})[1].span.text) # check_in 
        print ("check_in : " + soup_new.findAll("div",{"class":"amenitiesCategory"})[1].span.text)
    except : 
        check_in.append("nan")
        print ("check_in : nan")
        
    try:
        check_out.append(soup_new.findAll("div",{"class":"amenitiesCategory"})[1].findAll("span")[1].text) # check_out
        print ("check_out : " + soup_new.findAll("div",{"class":"amenitiesCategory"})[1].findAll("span")[1].text)
    except : 
        check_out.append("nan")
        print ("check_out : nan")
        
    try:
        room.append(soup_new.findAll("div",{"class":"amenitiesCategory"})[1].findAll("span")[2].text) # number of rooms
        print ("number of room : " +soup_new.findAll("div",{"class":"amenitiesCategory"})[1].findAll("span")[2].text)
    except :
        room.append("nan")
        print("room info not found")
    
    a = []
    for i in range(3,len(soup_new.findAll("div",{"class":"amenitiesCategory"}))):
        try:
            a.append(soup_new.findAll("div",{"class":"amenitiesCategory"})[i].text.replace('\t','').replace('\n',' '))
        except :
            a.append("nan")
            print("nan")
        
    hotel_info.append((a))
    
    print (hotel_info[l])
    
    
```

```python
print(len(hotel_name))
print(len(link))
print(len(num_review))
print(len(price))
print(len(rate))
print(len(check_in))
print(len(check_out))
print(len(addr))
print(len(room))
print(len(hotel_info))
```
